```python
import statsmodels.api as sm
X = df[['Digital (Million $)', 'radio (Million $)', 'TV (Million $)']]
y = df['sales (Million $)']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

Model Fit: The model explains ~87% of the variance in sales (R² = 0.869), indicating a strong fit.

Significant Predictors: Digital and Radio ad spend are statistically significant (p < 0.001), meaning they strongly impact sales.

Insignificant Predictor: TV ad spend has a very high p-value (0.478), indicating no significant effect on sales.

Intercept Meaning: When all ad spends are zero, expected base sales are ~$63M (interpreted from the constant term).

Q) US went through a tough period during COVID (2019 to 2020), for this analysis consider covid lockdowns placed in Q2-2020 and slowly relaxed in Q3-2020. Evaluate if COVID had an impact on sales and why ?

```python
sales_df['Timeframe'] = pd.to_datetime(sales_df['Timeframe'])
sales_df['Year'] = sales_df['Timeframe'].dt.year
sales_df['Quarter'] = sales_df['Timeframe'].dt.to_period('Q').astype(str).str.replace('Q', '- Q')
sales_2020 = sales_df[sales_df['Year'] == 2020]
quarterly_sales_2020 = sales_2020.groupby('Quarter')['sales (Million $)'].sum().reset_index()
```

```python
# Plot sales trend for 2020
plt.figure(figsize=(8, 5))
plt.plot(quarterly_sales_2020['Quarter'], quarterly_sales_2020['sales (Million $)'], marker='o', linestyle='-', color='teal')
plt.title('Quarterly Sales Trend in 2020 (COVID Impact)', fontsize=14)
plt.xlabel('Quarter')
plt.ylabel('Sales (Million $)')
plt.axvline(x='2020 - Q2', color='red', linestyle='--', label='Lockdown Starts')
plt.axvline(x='2020 - Q3', color='green', linestyle='--', label='Restrictions Ease')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The chart shows a sharp drop in sales in Q3 2020, aligning with the peak impact of COVID-19 lockdowns.
Sales decreased from ~$700M in Q2 to ~$540M in Q3, indicating disrupted consumer activity and restricted operations.
A partial rebound in Q4 2020 suggests recovery as restrictions eased.
This confirms that COVID-19 significantly impacted sales, especially during the strict lockdown period.
